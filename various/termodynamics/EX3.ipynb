# **Maxwell-Bolzmann distribution**

I generate random numbers and trasform them into range $[-1, 1]$. Then I combine three $erf^{-1}(y)$ functions to see, if they give me Gaussian distributions.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erfinv, gamma
```

```python
x = (np.random.randint(0, 1000, 1000) / 1000 - 0.5) * 2
y = (np.random.randint(0, 1000, 1000) / 1000 - 0.5) * 2
z = (np.random.randint(0, 1000, 1000) / 1000 - 0.5) * 2
x2 = (np.random.randint(0, 1000, 1000000) / 1000 - 0.5) * 2
y2 = (np.random.randint(0, 1000, 1000000) / 1000 - 0.5) * 2
z2 = (np.random.randint(0, 1000, 1000000) / 1000 - 0.5) * 2
mb = np.zeros(1000)
mb2 = np.zeros(1000000)

for i in range(len(x)):
    mb[i] = erfinv(x[i]) + erfinv(y[i]) + erfinv(z[i])

for i in range(len(x2)):
    mb2[i] = erfinv(x2[i]) + erfinv(y2[i]) + erfinv(z2[i])

plt.hist(mb, bins = 50, range = (-6, 6), color = 'teal', alpha = 0.5, lw=3, edgecolor='black', linewidth=1.2)
plt.show()
plt.hist(mb2, bins = 50, range = (-6, 6), color = 'orangered', alpha = 0.5, lw=3, edgecolor='black', linewidth=1.2)
plt.show()
```
