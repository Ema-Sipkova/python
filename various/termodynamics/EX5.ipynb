# **Calculation of Fermionic Integral**

```python
from scipy.integrate import quad
from scipy.special import gamma
import numpy as np
import matplotlib.pyplot as plt
from pynverse import inversefunc
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
def integral(y):
    return lambda x: x ** (1/2) / (np.exp(x - y) + 1)
```

# a)

```python
y = 0

Result = quad(integral(y), 0, np.inf)[0] / gamma(3/2)

print('F(0): {:.10f}'.format(Result))
```

# b)

```python
def f(y):
    I = lambda x, y: x ** (1/2) / (np.exp(x - y) + 1)
    return quad(I, 0, np.inf, (y))[0] / gamma(3/2)

inv = inversefunc(f)
for Y in [80, 10, 0.1]:
	print("ni_{0} = {1:.4f}".format(Y, inv(Y)))
```

# c)

```python
def N(e, ni):
	return (np.exp(e-ni) + 1)**(-1)

E = np.linspace(-30, 40, 1000)
for Y in [100, 1.0, 0.01]:
	plt.plot(E, N(E, inv(Y)), label="$x\,=\,${0}, $\mu_x\,=\,${1:.3f}".format(Y, inv(Y)))
plt.xlabel("$\epsilon$")
plt.ylabel("$N$")
plt.legend()
plt.show()
```
